```python
import pandas as pd
from joint_rpca_functions import ord_scatter_3D

%matplotlib inline
```

```python
#load site specific results
path_ = '../../../../Code/Joint-RPCA/output/Site_level_PCA_output.csv'
site_pca = pd.read_csv(path_, index_col=0).dropna()

new_labels = {"MISAME-3": "MISAME", "VITAL-Lactation": "Mumta-LW", "CHILD": "CHILD"}
site_pca['STUDYID_NEW'] = site_pca.STUDYID.map(new_labels)
```

```python
site_palette = {"MISAME": '#95AAD3', "Mumta-LW": '#94C47D', "CHILD": '#EDAEAE'}

ord_scatter_3D(ord_dict={"": site_pca}, 
                hue='STUDYID_NEW',
                hue_order=site_palette.keys(), 
                legend_title='', pcs=["PC1","PC2","PC3"],
                palette=site_palette.values(),
                subplots=None, 
                figsize=(8, 8), elev=20, axim=65, save_fig=False,
                save_path="../../../../Figures/Joint-RPCA/Supplementary/STUDY_PCA.pdf")
```
